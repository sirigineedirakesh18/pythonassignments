```python
def fibonacci(k):
    if k>1:
        return fibonacci(k-1)+fibonacci(k-2)
    else:
        return k
k=int(input("enter sequence terms: "))
for i in range(k):
    print(fibonacci(i))
```

```python
def fact(n):
    if n==0 or n==1:
        return 1
    elif n<0:
        return "enter positive number"
    else:
        return n*fact(n-1)
print(fact(4))
    
    
```

```python
weight=float(input("enter weight in kg : "))
height=float(input("enter height in cms : "))
bmi=weight/(height)**2
print(round(bmi,2))
if bmi<18.5:
    print("underweight")
elif bmi>25:
    print("overweight")
else:
    print("normal")
```
