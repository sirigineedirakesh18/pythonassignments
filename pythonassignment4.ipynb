```python
#incase of normal list shallowcopy==deepcopy but for nested list
```

```python
#shallow copy
import copy
l1=[[1,2,3],[4,5,6]]
l2=copy.copy(l1)
l1,l2
```

```python
l2[1][0]=100
l1,l2
```

```python
#deepcopy
l1=[[1,2,3],[4,5,6]]
l2=copy.deepcopy(l1)
l1,l2
```

```python
l2[1][0]=100
l1,l2
```
