```python
a=int(input("Enter First Number: "))
b=int(input("Enter Second Number: "))
if a>b:
    big=a
else:
    big=b
while(True):
    if(big%a==0)&(big%b==0):
        lcm=big
        break
    big=big+1
print("LCM IS",lcm)
```

```python
a=int(input("Enter First Number: "))
b=int(input("Enter Second Number: "))
if a<b:
    small=a
else:
    small=b
while(True):
    if(a%small==0)&(b%small==0):
        hcf=small
        break
    small=small-1
print("HCF IS",hcf)
```

```python
a=int(input("Enter First Number: "))
b=int(input("Enter Second Number: "))
what_to_do=input("enter what to do ")
if what_to_do=="add":
    print(a+b)
elif what_to_do=="sub":
    print(a-b)
elif what_to_do=="mul":
    print(a*b)
elif what_to_do=="div":
    print(a/b)
else:
    print("please enter correct value")
```
