```python
5. Write a Python Program to Find Armstrong Number in an Interval
```

```python
n=int(input("starting interval: "))
o=int(input("ending interval: "))
m=list(range(n,o))
l1=[]
l2=[]
for i in range(n,o):
    a=[int(j) for  j in str(i)]
    l=sum(list(map(lambda b:b**3,a)))
    l1.append(l)
for i in range(len(l1)):
    if l1[i]==m[i]:
        l2.append(l1[i])
print(l2)
    
```
